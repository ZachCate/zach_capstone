```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from ipywidgets import interact
import plotly.express as px
import statsmodels.api as sm
```

```python
non_country_values = ['Asia', 'high-income countries', 'Upper-middle-income countries', 'World',
                      'North America', 'Europe', 'Africa', 'European Union', 'South America',
                      'Lower-middle-income countries', 'low-income countries', 'European Union (27)', 'European Union (27) (GCP)', 'High-income countries']
```

```python
owid = owid[~owid['country'].isin(non_country_values)]
```

```python
def update_chart(year):
    data_year = owid[owid['year'] == year]
    top_countries = data_year.nlargest(10, 'co2')['country']
    top_co2 = data_year[data_year['country'].isin(top_countries)]

    plt.figure(figsize=(10, 6))
    plt.bar(top_co2['country'], top_co2['co2'], color='skyblue')
    plt.xlabel('Country')
    plt.ylabel('CO2 Emissions (Million Metric Tons)')
    plt.title(f'Top 10 CO2 Emitting Countries in {year}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

years = owid['year'].unique()

interact(update_chart, year=years)
```

```python
def update_scatter_tooltip(year_selected):
    data_selected_year = owid[owid['year'] == year_selected]

    fig = px.scatter(data_selected_year, x='primary_energy_consumption', y='co2', color='country', hover_name='country')
    fig.update_layout(
        title=f'Relationship between Primary Energy Consumption and CO2 Emissions in {year_selected}',
        xaxis_title='Primary Energy Consumption (Gigajoules per capita)',
        yaxis_title='CO2 Emissions (Million Metric Tons)',
        template='plotly_dark'
    )
    fig.show()

unique_years = owid['year'].unique()

interact(update_scatter_tooltip, year_selected=unique_years)
```

```python
def update_bar_chart_simple(year_selected):
    data_selected_year = owid[owid['year'] == year_selected]
    data_sources = data_selected_year[['coal_co2', 'oil_co2', 'gas_co2', 'flaring_co2', 'cement_co2', 'land_use_change_co2']]

    sources = ['Coal', 'Oil', 'Gas', 'Flaring', 'Cement', 'Land Use Change']
    emissions = data_sources.values.tolist()[0]

    plt.figure(figsize=(10, 6))
    plt.bar(sources, emissions, color='purple')
    plt.xlabel('Source')
    plt.ylabel('CO2 Emissions (Million Metric Tons)')
    plt.title(f'CO2 Emissions from Different Sources in {year_selected}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

unique_years = owid['year'].unique()

interact(update_bar_chart_simple, year_selected=unique_years)
```

```python
def update_scatter_primary_energy_new(year_selected):
    data_selected_year = owid[owid['year'] == year_selected]

    fig = px.scatter(data_selected_year, x='primary_energy_consumption', y='co2', color='country', hover_name='country')
    fig.update_layout(
        title=f'Relationship between Primary Energy Consumption and CO2 Emissions in {year_selected}',
        xaxis_title='Primary Energy Consumption (Gigajoules per capita)',
        yaxis_title='CO2 Emissions (Million Metric Tons)',
        template='plotly_dark'
    )
    fig.show()

unique_years_scatter = owid['year'].unique()

interact(update_scatter_primary_energy_new, year_selected=unique_years_scatter)
```
