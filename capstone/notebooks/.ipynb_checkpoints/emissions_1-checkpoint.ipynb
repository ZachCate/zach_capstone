```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from ipywidgets import interact
import plotly.express as px
import statsmodels.api as sm
import os
```

```python
print(os.getcwd())
```

```python
world_emissions = pd.read_csv('..\\..\\capstone\\data\\world_emissions.csv')
world_emissions.head()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=world_emissions, x='Year', y='Total (billion $2019)', hue='Region')
plt.title('Total Energy Consumption Over Years')
plt.show()
```

```python
correlation_matrix = world_emissions.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
world_emissions['Non-Fossil'] = world_emissions['Renewables'] + world_emissions['Electricity networks'] + world_emissions['Other supply'] + world_emissions['Low-carbon fuels']

grouped_data = world_emissions.groupby(['Year'])['Non-Fossil'].sum().reset_index()

plt.figure(figsize=(12, 8))
sns.set_palette("Set3")
sns.lineplot(data=grouped_data, x='Year', y='Non-Fossil', marker="o")

plt.title('Non-Fossil Energy Use Over Years')
plt.xlabel('Year')
plt.ylabel('Non-Fossil Energy Consumption')
plt.show()
```

```python
correlation = world_emissions['Non-Fossil'].corr(world_emissions['Energy efficiency'])
print("Correlation Coefficient:", correlation)
```

```python
plt.figure(figsize=(10, 6))
sns.set_palette("Set3")

sns.regplot(data=world_emissions, x='Non-Fossil', y='Energy efficiency', scatter_kws={'alpha': 0.7, 'label': world_emissions['Region']})
plt.title('Correlation between Non-Fossil Energy Adoption and Energy Efficiency')
plt.xlabel('Non-Fossil Energy Adoption')
plt.ylabel('Energy Efficiency')
plt.legend()
plt.show()
```

```python
unique_regions = world_emissions['Region'].unique()

for region in unique_regions:
    plt.figure(figsize=(10, 6))
    sns.set_palette("Set3")
    
    region_data = world_emissions[world_emissions['Region'] == region]
    
    sns.regplot(data=region_data, x='Non-Fossil', y='Energy efficiency', scatter_kws={'alpha': 0.7}, label=region)
    plt.title(f'Correlation between Non-Fossil Energy Adoption and Energy Efficiency in {region}')
    plt.xlabel('Non-Fossil Energy Adoption')
    plt.ylabel('Energy Efficiency')
    plt.legend()
    plt.show()
```

```python
coal = pd.read_csv('..\\..\\capstone\\data\\coal.csv')
oil = pd.read_csv('..\\..\\capstone\\data\\oil.csv')
gas = pd.read_csv('..\\..\\capstone\\data\\gas.csv')
```

```python
oil.tail()
```

```python
gas.shape
```

```python
coal['fuel type'] = 'coal'
gas['fuel type'] = 'gas'
oil['fuel type'] = 'oil'
```

```python
coal.columns = coal.columns.str.strip()
gas.columns = gas.columns.str.strip()
oil.columns = oil.columns.str.strip()
```

```python
fossil_fuels = pd.concat([coal, gas, oil])
```

```python
fossil_fuels.reset_index(drop=True, inplace=True)
```

```python
fossil_fuels.tail()
```

```python
fossil_fuels.shape
```

```python
fossil_fuels.dtypes
```

```python
plt.figure(figsize=(12, 6))

fuel_types = ['coal', 'oil', 'gas']

for fuel_type in fuel_types:
    fuel_data = fossil_fuels[fossil_fuels['fuel type'] == fuel_type]
    plt.plot(fuel_data['year'], fuel_data['World'], marker='o', label=fuel_type.capitalize())

plt.xlabel('Year')
plt.ylabel('CO2 Emissions')
plt.title('CO2 Emissions by Fuel Type')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
